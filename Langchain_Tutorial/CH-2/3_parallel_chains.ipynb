```python
# TASK-3  StrParser

str_parser = StrOutputParser()
```

```python
# TASK-4  [Custom Runnable]  
from langchain_core.runnables import RunnableLambda

def dictionary_maker(text: str) -> dict:

 return{"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)
```

### **parallel CHAIN 1**

```python
# TASK - 1 [prompt]

linked_prompt = ChatPromptTemplate.from_messages(
[("system", "你是一个微博帖子生成器"),
("human", "给我生成一个关于{text}的微博帖子")])

# TASK - 2 [LLM] llm_openai = ChatOpenAI(model = "deepseek-chat",

# TASK - 3 [StrParser] str_parser = StrOutputParser()

chain_linked = linked_prompt | llm_openai | str_parser
```

### **parallel CHAIN 2**

```python
from langchain_core.runnables import RunnableParallel
```

```python
def insta_chain(text: dict) :

 text = text["text"]

 # TASK - 1 [prompt]
 insta_prompt = ChatPromptTemplate.from_messages(
     [("system", "你是一个小红书帖子生成器"),
     ("human", "给我生成一个关于{text}的小红书的帖子")])
 # TASK - 2 [LLM] llm_openai = ChatOpenAmodel =     "deepseek-chat",
 # TASK - 3 [StrParser] str_parser =trOutputParser()

 chain_insta = insta_prompt | llm_openai | str_parser

 result =  chain_insta.invoke(text)

 return result

insta_chain_runnable = RunnableLambda(insta_chain)
```

### **Final Orchestration**

```python
final_chain =(
   prompt_template | llm_openai |    str_parser | dictionary_maker_runnable | RunnableParallel({
       "weibo": chain_linked,
       "insta": insta_chain_runnable
   })
   )
```

```python
final_chain.invoke("千与千寻")
```

### **Chain as a Runnable**

```python
# TASK - 1 [Beautify Function]

def beautify(final_response: dict) -> dict:

 weibo_response = final_response["weibo"]
 insta_response = final_response["insta"]

 return {
       "weibo": f"""微博帖子：
{weibo_response}

""",
        
        "insta": f"""小红书帖子：
{insta_response}
"""
 }

# TASK - 2 [Runnable]

beautify_runnable = RunnableLambda(beautify)

# TASK - 3 [Final Chain]

beautify_chain = final_chain | beautify_runnable
beautify_chain.invoke("千与千寻")
```
