# **MESSAGE**

```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

if os.environ.get('DEEPSEEK_API_KEY') :
    print("DEEPSEEK_API_KEY is set")
else:    print("DEEPSEEK_API_KEY is not set")


```

```python
from langchain_core.messages import HumanMessage, SystemMessage,AIMessage
from langchain_openai import ChatOpenAI

my_messages = [
   HumanMessage(content="Bro ,告诉我一个冷知识"),
    SystemMessage(content="你是一个有趣的助手"),
    ]

llm_openai = ChatOpenAI(
    api_key=os.environ.get('DEEPSEEK_API_KEY'),
        base_url="https://api.deepseek.com",
    model="deepseek-chat",
    temperature=0,
)

llm_openai.invoke(my_messages).content
```

# **Prompts**



```python
from langchain_core.messages import HumanMessage, SystemMessage,AIMessage
from langchain_openai import ChatOpenAI
import os


llm_openai = ChatOpenAI(model = "deepseek-chat",
base_url="https://api.deepseek.com",
api_key=os.environ.get('DEEPSEEK_API_KEY'),
 temperature=0)
my_messages = [
   HumanMessage(content="Bro ,告诉我一个冷知识"),
    SystemMessage(content="你是一个有趣的助手"),
    ]

llm_openai.invoke(my_messages).content   
```

```python
from langchain_core.prompts import PromptTemplate

user_input = input("请输入一个话题：")

dynamic_prompt = PromptTemplate.from_template("Bro ,告诉我一个{topic}的冷知识")

ready_prompt = dynamic_prompt.invoke({"topic": user_input })

llm_openai.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
 ("system", "你是一个{tone}的助手"),
 ("user", "Bro ,告诉我一个{topic}的冷知识")
])

user_input = input("请输入一个话题：")
tone_input = input("请输入一个语气：")

ready_prompt = prompt_template.invoke({"topic": user_input, "tone": tone_input })

llm_openai.invoke(ready_prompt.messages).content
```
