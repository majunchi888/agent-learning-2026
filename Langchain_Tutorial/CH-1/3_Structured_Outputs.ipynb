```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv
from langchain_core.messages import HumanMessage, SystemMessage,AIMessage
from langchain_openai import ChatOpenAI
import os

# Load environment variables from .env file
load_dotenv()

llm_openai = ChatOpenAI(model = "deepseek-chat",
base_url="https://api.deepseek.com",
api_key=os.environ.get('DEEPSEEK_API_KEY'),
 temperature=0)


if os.environ.get('DEEPSEEK_API_KEY') :
    print("DEEPSEEK_API_KEY is set")
else:    print("DEEPSEEK_API_KEY is not set")


```

```python
from langchain_core.prompts import PromptTemplate

llm_openai.invoke("Bro ,告诉我一个生理冷知识，以键值对形式回答").content
```

# **Using Pydantic Models**

```python
from pydantic import BaseModel,Field

class llm_schema(BaseModel):
     setup: str = Field(description="The setup of the joke")
     punchline: str = Field(description="The punchline of the joke")
```

```python
llm_structured_output = llm_openai.with_structured_output(llm_schema,method="function_calling"  # 非openai模型需要指定method="function_calling" 
)

llm_structured_output.invoke("Tell me a joke ")
```

# **Using TypeDict (not strict)**

```python
from typing import TypedDict

class llm_schema_td(TypedDict):
    setup: str
    punchline: str
```

```python
llm_structured_typed_dict = llm_openai.with_structured_output(llm_schema_td,method="function_calling" ) # 非openai模型需要指定method="function_calling")

llm_structured_typed_dict.invoke("Tell me a joke ")
```
